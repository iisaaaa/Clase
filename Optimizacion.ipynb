**El problema**

Se requiere encontrar los valores extremos (máximo y mínimo) de una función $f$ en un intervalo $I$.
$$$$

**Motivacion**

Encontrar los extremos de una función permite resolver ciertos tipos de problemas de optimización.
$$$$

**Definicion**

$i)$ Un número $f(c)$ es un máximo absoluto de una función $f$ si para todo $x$ en el dominio de $f$, $~$ $f(x) \leq f(c)$.

$ii)$ Un número $f(c)$ es un mínimo absoluto de una función $f$ si para todo $x$ en el dominio de $f$, $~$ $f(x) \geq f(c)$.
$$$$



image.png


**Definición**

Si $𝑓$ es diferenciable en $𝑐$ y $𝑓′(𝑐) = 0$, entonces llamamos a $𝑐$ un punto crítico o punto estacionario de $𝑓$. Un punto $𝑐$ en el que la derivada de $𝑓$ no está definida se llama punto singular de $𝑓$.
$$$$

De acuerdo con esto, los candidatos para la ubicación de los valores extremos de una función continua en un intervalo cerrado se dividen en tres categorías: $(a)$ puntos limite del intervalo, $(b)$ puntos críticos y $(c)$ puntos singulares. Para determinar los valores extremos de una función $f$ identificamos estos puntos, evaluamos la funcion en dichos puntos y determinamos el mayor y el menor valor de la función.

**Ejemplo**

Encontrar los valores máximo y mínimo de la función
$$$$

$$g(t)=cos(t)+sin(t)$$
$$$$

En el intervalo $[0,2\pi]$
$$$$


**Solucion**

Hallamos los puntos críticos
$$$$

$$g'(t)=-sin(t)+cos(t)$$

$$0=-sin(t)+cos(t)$$

$$sin(t)=cos(t)$$
$$$$

Para resolver esta ecuación se divide ambos lados por $cos(t)$ y se encuentra
$$$$

$$tan(t)=1$$
$$$$

Lo cual, en el intervalo considerado implica $t=\frac{\pi}{4}$, o $t=\frac{5\pi}{4}$. Puesto que no hay puntos singulares, evaluamos $g(t)$ en los puntos límite del intervalo y en los puntos críticos.
$$$$

$$g(0)=1+0=1$$

$$g(\frac{\pi}{4})=\frac{1}{\sqrt2}+\frac{1}{\sqrt2}=\sqrt2$$

$$g(\frac{5\pi}{4})=-\frac{1}{\sqrt2}-\frac{1}{\sqrt2}=-\sqrt2$$

$$g(2\pi)=1+0=1$$
$$$$

Asi que $g(t)$ tiene un valor máximo de $\sqrt2$ en $t=\frac{\pi}{4}$ y un valor mínimo de $-\sqrt2$ en $t=\frac{5\pi}{4}$.

**Extremos locales**

Si estudiamos el comportamiento de funciones en intervalos abiertos no podemos hablar de valores extremos absolutos, debemos hablar de valores extremos locales, como se vera en la siguiente proposicion.
$$$$

**Proposición**

Supongamos que 𝑓 y 𝑓′ son continuas en un intervalo abierto (𝑎, 𝑏) y 𝑐 es el único punto crítico de 𝑓 en (𝑎, 𝑏). Si  𝑓(𝑐) es el valor mínimo de 𝑓 en (𝑎, 𝑏) entonces 𝑓 tiene un mínimo local en 𝑐; si 𝑓 es el valor máximo de 𝑓 en (𝑎, 𝑏) entonces 𝑓  tiene un máximo local en 𝑐.
​$$$$

**Proposición**

Supongamos que 𝑓, 𝑓′ y 𝑓″ son continuas en un intervalo abierto (𝑎, 𝑏) y que 𝑐 es un punto crítico de 𝑓 en (𝑎, 𝑏). Entonces 𝑓 tiene un máximo local en 𝑐 si 𝑓″(𝑐) < 0 y 𝑓 tiene un mínimo local en 𝑐 si 𝑓″(𝑐) > 0.
$$$$

```python
from sympy import *
from sympy.plotting import plot
from sympy.abc import x
init_printing()
f= 5*x**3-3*x**5
def maxminf(f):
    """ Calcula los máximos y mínimos de una función f(x) """
    df = diff(f, x) # 1era. derivada
    d2f = diff(f, x, 2) # 2da. derivada
    pcs = solve(Eq(df,0)) # puntos críticos
    for p in pcs:
        if d2f.subs(x,p)>0:
            tipo="Min"
        elif d2f.subs(x,p)<0:
            tipo="Max"
        else:
            tipo="Indefinido"
        print("x = %f (%s)"%(p,tipo))
```

```python
from sympy import *
from sympy.plotting import plot
init_printing()

def maxminf(f, var, intervalo):
    """ Calcula máximos y mínimos de una función f(var) en un intervalo cerrado """
    print(f"\nFunción: f({var}) = {f}")
    df = diff(f, var)
    d2f = diff(f, var, 2)
    print(f"Primera derivada: {df}")

    pcs = solve(Eq(df, 0), var)
    extremos = []

    for p in pcs:
        if p.is_real and intervalo[0] <= p.evalf() <= intervalo[1]:
            concavidad = d2f.subs(var, p)
            if concavidad > 0:
                tipo = "Min"
            elif concavidad < 0:
                tipo = "Max"
            else:
                tipo = "Indefinido"
            extremos.append((p.evalf(), f.subs(var, p).evalf(), tipo))

    # Evaluamos extremos del intervalo
    extremos.append((intervalo[0], f.subs(var, intervalo[0]).evalf(), "Extremo (borde)"))
    extremos.append((intervalo[1], f.subs(var, intervalo[1]).evalf(), "Extremo (borde)"))

    for x_val, y_val, tipo in extremos:
        print(f"{var} = {x_val}, f({var}) = {y_val} ({tipo})")

    p = plot(f, (var, intervalo[0], intervalo[1]), show=False, line_color='blue')
    p.title = f"f({var}) = {f}"
    p.xlabel = str(var)
    p.ylabel = f"f({var})"
    p.show()

x1 = symbols('x')
f1 = x1**2 - 4
maxminf(f1, x1, (-3, 4))
```

(e) $g(x) = x^2 \cos(x)$ on $[-2,2]$

```python
from sympy import *
from sympy.plotting import plot
init_printing()

def maxminf(f, var, intervalo, puntos=100):
    """ Calcula máximos y mínimos de una función f(var) en un intervalo cerrado usando nsolve directamente """
    print(f"\nFunción: f({var}) = {f}")
    df = diff(f, var)
    d2f = diff(f, var, 2)
    print(f"Primera derivada: {df}")

    # Usamos búsqueda numérica en una malla
    pcs = []
    paso = (intervalo[1] - intervalo[0]) / puntos
    for i in range(puntos):
        guess = intervalo[0] + i * paso
        try:
            root = nsolve(df, var, guess)
            root_val = N(root)
            if intervalo[0] <= root_val <= intervalo[1]:
                # Evitamos duplicados
                if all(abs(root_val - pc) > 1e-4 for pc in pcs):
                    pcs.append(root_val)
        except:
            continue

    extremos = []

    for p in pcs:
        concavidad = d2f.subs(var, p).evalf()
        if concavidad > 0:
            tipo = "Min"
        elif concavidad < 0:
            tipo = "Max"
        else:
            tipo = "Indefinido"
        extremos.append((p, f.subs(var, p).evalf(), tipo))

    # Evaluamos extremos del intervalo
    extremos.append((intervalo[0], f.subs(var, intervalo[0]).evalf(), "Extremo (borde)"))
    extremos.append((intervalo[1], f.subs(var, intervalo[1]).evalf(), "Extremo (borde)"))

    for x_val, y_val, tipo in extremos:
        print(f"{var} = {x_val:.5f}, f({var}) = {y_val:.5f} ({tipo})")

    # Gráfica
    p = plot(f, (var, intervalo[0], intervalo[1]), show=False, line_color='blue')
    p.title = f"f({var}) = {f}"
    p.xlabel = str(var)
    p.ylabel = f"f({var})"
    p.show()



x3 = symbols('x')
f5 = x3**2 * cos(x3)
maxminf(f5, x3, (-2, 2))
```

```python
from sympy import *
from sympy.plotting import plot
init_printing()

def maxminf(f, var, intervalo, puntos=100):
    """ Calcula máximos y mínimos de una función f(var) en un intervalo cerrado usando nsolve directamente """
    print(f"\nFunción: f({var}) = {f}")
    df = diff(f, var)
    d2f = diff(f, var, 2)
    print(f"Primera derivada: {df}")

    # Usamos búsqueda numérica en una malla
    pcs = []
    paso = (intervalo[1] - intervalo[0]) / puntos
    for i in range(puntos):
        guess = intervalo[0] + i * paso
        try:
            root = nsolve(df, var, guess)
            root_val = N(root)
            if intervalo[0] <= root_val <= intervalo[1]:
                # Evitamos duplicados
                if all(abs(root_val - pc) > 1e-4 for pc in pcs):
                    pcs.append(root_val)
        except:
            continue

    extremos = []

    for p in pcs:
        concavidad = d2f.subs(var, p).evalf()
        if concavidad > 0:
            tipo = "Min"
        elif concavidad < 0:
            tipo = "Max"
        else:
            tipo = "Indefinido"
        extremos.append((p, f.subs(var, p).evalf(), tipo))

    # Evaluamos extremos del intervalo
    extremos.append((intervalo[0], f.subs(var, intervalo[0]).evalf(), "Extremo (borde)"))
    extremos.append((intervalo[1], f.subs(var, intervalo[1]).evalf(), "Extremo (borde)"))

    for x_val, y_val, tipo in extremos:
        print(f"{var} = {x_val:.5f}, f({var}) = {y_val:.5f} ({tipo})")

    # Gráfica
    p = plot(f, (var, intervalo[0], intervalo[1]), show=False, line_color='blue')
    p.title = f"f({var}) = {f}"
    p.xlabel = str(var)
    p.ylabel = f"f({var})"
    p.show()


t4 = symbols('t')
f7 = t4**2 * sin(t4)
maxminf(f7, t4, (0, pi))
```

(11) A farmer wishes to construct a rectangular storage bin with a volume of $100$ cubic feet. Both the top and the bottom of the bin are to be squares. find the dimensions of the bien which will minimize its surface area.

(11) Un agricultor desea construir un contenedor rectangular con un volumen de $100$ pies cúbicos. Tanto la parte superior como la inferior deben ser cuadradas. Determine las dimensiones del contenedor que minimicen su superficie.

- Variables y restricciones:
  - Supongamos que:
    - $x$ es el lado de la base cuadrada
    - $h$ es la altura de la caja
  - Como el volumen es $1000$:
  $$x^2h = 1000$$ $(1)$

  - El área superficial (que queremos minimizar) es:
  $$A(x,h) = 2x^2 + 4xh$$ $(2)$

  ($2x^2$ por las dos tapas cuadradas (base y superior), y $4xh$ por los lados rectangulares)

- Sustituimos $h$ en términos de $x$:
  - De $(1)$:
  $$h = \frac{1000}{x^2}$$
  - Sustituimos en $(2)$:
  $$A(x) = 2x^2 + 4x * \frac{1000}{x^2} = 2x^2 + \frac{4000}{x}$$

- Derivada y mínimo:

$$A'(x) = 4x - \frac{4000}{x^2}$$

Buscamos $A'(x) = 0$:
$$4x - \frac{4000}{x^2} = 0 → 4x^3 = 4000 → x^3 = 1000 → x = 10 $$

$$h = \frac{1000}{10^2} = \frac{1000}{100} = 10$$

- Resultado:

  - Lado de base: $x = 10$ pies
  - Altura: $h = 10$ pies
  - Es un cubo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Volumen constante
V = 1000

# Función del área total
def A(x):
    return 2 * x**2 + 4000 / x

# Derivada e investigación de mínimo
x_vals = np.linspace(1, 20, 400)
A_vals = A(x_vals)

min_x = 10  # Como se calculó antes
min_h = V / (min_x**2)

plt.plot(x_vals, A_vals, label='Área total')
plt.axvline(min_x, color='red', linestyle='--', label='x óptimo')
plt.xlabel('x (pies)')
plt.ylabel('Área superficial (ft²)')
plt.title('Ejercicio 11: Minimizar área total')
plt.legend()
plt.grid()
plt.show()

print(f"Lado óptimo: {min_x:.2f} ft, Altura: {min_h:.2f} ft")
```

(17) Show that the rectangle with meximum area for a givenj perimeter $P$ is a square.

(17) Demuestre que el rectángulo con área máxima para un perímetro j dado $P$ es un cuadrado.

Perimetro $P = 2x + 2y ⇒ y = \frac{P}{2} - x$

$A = x *y = x (\frac{P}{2} - x) = -x^2 + \frac{P}{2}x$

Máxima área:

$$
A'(x) = -2x + \frac{P}{2} = 0 ⇒ x = \frac{P}{4}, ~~ y = \frac{P}{4} ⇒ Es~~ un~~ cuadrado
$$  

Conclusión: El rectángulo con área máxima para un perímetro fijo es un cuadrado.

```python
# Demostración simbólica para área máxima con perímetro fijo
from sympy import symbols, Eq, solve, diff, simplify

x, P = symbols('x P')
y = P/2 - x
A = x * y
dA = diff(A, x)

crit = solve(dA, x)[0]
y_val = y.subs(x, crit)
simplified_A = simplify(A)

print("Área máxima cuando:")
print(f"x = y = {crit} → Cuadrado")
print(f"Área en términos de x: {simplified_A}")
```
